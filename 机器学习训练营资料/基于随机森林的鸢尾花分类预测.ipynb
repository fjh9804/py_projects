## 随机森林
随机森林是一种监督式算法，使用由众多决策树组成的一种集成学习方法，输出是对问题最佳答案的共识。随机森林可用于分类或回归。

### 随机森林API

- class sklearn.ensemble.RandomForestClassifier(n_estimators=10, criterion=’gini’, max_depth=None, bootstrap=True, random_state=None, min_samples_split=2)

    - 随机森林分类器
    - n_estimators：integer，optional（default = 10）森林里的树木数量120,200,300,500,800,1200
    - criteria：string，可选（default =“gini”）分割特征的测量方法
    - max_depth：integer或None，可选（默认=无）树的最大深度 5,8,15,25,30
    - max_features="auto”,每个决策树的最大特征数量
        - If "auto", then max_features=sqrt(n_features).
        - If "sqrt", then max_features=sqrt(n_features) (same as "auto").
        - If "log2", then max_features=log2(n_features).
        - If None, then max_features=n_features.
    - bootstrap：boolean，optional（default = True）是否在构建树时使用放回抽样
    - min_samples_split:节点划分最少样本数
    - min_samples_leaf:叶子节点的最小样本数
- 超参数：n_estimator, max_depth, min_samples_split,min_samples_leaf

### 随机森林项目实战：鸢尾花分类 🌳🌳🌳

**目标：**用随机森林判断鸢尾花的品种（山鸢尾、变色鸢尾、维吉尼亚鸢尾）

**工具：**Python的sklearn库（代码仅需10行）

鸢【音：yuān】尾花（Iris）是单子叶百合目花卉，是一种比较常见的花，可能不经意间你就能在某个公园里碰见它，而且鸢尾花的品种较多。



鸢尾花（iris）数据集，它共有4个属性列和一个品种类别列：sepal length（萼片长度）、sepal width（萼片宽度）、petal length（花瓣长度）、petal width （花瓣宽度），单位都是厘米。3个品种类别是Setosa、Versicolour、Virginica，样本数量150个，每类50个。

### 导包

```python
from sklearn.datasets import load_iris # 加载数据
from sklearn.ensemble import RandomForestClassifier  # 随机森林的分类算法
from sklearn.model_selection import train_test_split # 数据集的划分（训练和测试）
```

### 加载数据

```python
iris = load_iris()
print(iris)
X = iris.data  # 特征
y = iris.target  # 标签

# 数据集的划分
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=200, max_depth=4)  # 创建随机森林算法（赋予学习能力）
model.fit(X_train, y_train)  # 学习训练集
```

```python
predictions = model.predict(X_test) # 考试
predictions  # 预测结果
```

```python
y_test  # 参考答案
```

```python
acc = model.score(X_test, y_test)
acc
```

### 特征重要性分析

```python
import matplotlib.pyplot as plt
print(model.feature_importances_)  # 打印特征重要性分数
plt.barh(iris.feature_names, model.feature_importances_)
```

### 关键概念解释

- 为什么叫「随机」？:每棵树训练时：① 随机选部分样本，② 随机选部分特征。避免所有树都一样，降低过拟合风险。
- 为什么比单棵决策树好？:单棵树容易「钻牛角尖」（过拟合），而森林通过投票排除个别树的错误。


## 基于随机森林的乳腺癌预测与可视化
### 第一部分：环境准备与数据加载
#### 步骤1：导入工具库

"这些工具库就像医生的装备箱：

- numpy 是数学计算的手术刀，处理各种数值运算

- pandas 是病例整理台，轻松处理表格数据

- matplotlib和seaborn是诊断影像设备，把数据变成直观图表"

```python
# ----------------- 1. 环境准备 -----------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import (accuracy_score, confusion_matrix,
                            classification_report, roc_curve, auc)
from sklearn.preprocessing import StandardScaler
```

#### 步骤2：加载数据

"这里加载的是威斯康星乳腺癌数据集：

- 包含569个病例（212恶性，357良性）

- 30个细胞特征：比如细胞半径、纹理、周长等

- 目标变量y是诊断结果：0是良性，1是恶性"

（展示数据前3行）
"大家看这些特征名称：

- 'mean radius'平均半径：癌细胞通常更大

- 'worst texture'最差纹理：表面最粗糙的区域

- 'area error'面积误差：不同切片测量的差异

这些特征就像肿瘤的'指纹'，帮助模型识别病变"

```python
# ----------------- 2. 数据加载与探索 -----------------
# 加载数据集
'''
569 (212恶性、357良性)
30 个细胞特征
0：良性、1：恶性
'''
data = load_breast_cancer()
print(data)
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

### 第二部分：数据探索
#### 步骤1：类别分布可视化

"这张条形图显示：

- 良性病例357个（蓝色）

- 恶性病例212个（橙色）

说明数据存在类别不平衡，但不算严重，暂时不需要特殊处理"

```python
# 创建数据概览可视化
plt.figure(figsize=(12, 8))
sns.countplot(x=y)
plt.title("Class Distribution (0: Benign, 1: Malignant)")
plt.show()
```

#### 步骤2：特征分布直方图

"我们查看前6个特征的分布：

- 'mean radius'呈双峰分布 → 可能对应良恶性差异

- 'mean texture'近似正态分布 → 正常细胞变异范围

- 'mean perimeter'与半径高度相关 → 后续可以研究特征相关性"

```python
# 特征分布直方图
plt.figure(figsize=(15, 10))
X.iloc[:, :6].hist(bins=20, figsize=(15, 10), layout=(2, 3))
plt.suptitle("Feature Distribution Histograms")
plt.tight_layout()
plt.show()
```

### 第三部分：数据预处理
#### 步骤1：数据标准化
"这步操作就像给所有特征做体检标准化：

- 把不同量纲的特征（比如半径单位是微米，纹理是无量纲值）统一成标准分

- 计算公式：(原始值 - 平均值)/标准差

- 为什么要标准化？避免数值范围大的特征（如面积）主导模型判断"

```python
# ----------------- 3. 数据预处理 -----------------
# 标准化处理
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### 步骤2：数据集划分
"这里把数据分成：

- 训练集455例（80%）：用于培养'实习医生'

- 测试集114例（20%）：最后的执业考试

random_state=42就像固定考试题目，确保每次划分结果一致"

```python
# 划分数据集
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)
```

### 第四部分：模型训练（15分钟）
#### 步骤1：初始化随机森林
"随机森林就像专家会诊：

- 每棵决策树是一个专科医生

- max_depth=5限制每个医生的经验层级（防止过拟合）

- n_estimators=100邀请100位专家投票诊断"

```python
# ----------------- 4. 模型训练 -----------------
# 初始化随机森林
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    random_state=42
)
```

#### 步骤2：交叉验证
"这是模拟执业考试：

- 把训练集分成5份，轮流用4份训练，1份验证

- 5次的平均准确率约0.95 → 实习医生表现优秀

- 为什么要交叉验证？避免运气好抽到简单题目"

```python
# 交叉验证评估
cv_scores = cross_val_score(rf, X_train, y_train, cv=5)  # 5折交叉验证
print(f"Cross-Validation Scores: {cv_scores}")
print(f"Mean CV Accuracy: {np.mean(cv_scores):.2f}")
```

#### 步骤3：正式训练
"这个过程就像：

- 每个决策树专家学习不同病例子集

- 学习细胞特征与诊断结果的关系

- 最终形成集体诊断规则"

```python
# 训练模型
rf.fit(X_train, y_train)
```

### 第五部分：模型评估
#### 步骤1：测试集评估
"最终考试成绩96分！说明：

- 在114个新病例中正确诊断了约109例

- 但要注意：准确率高不代表模型完美，还需看具体错误类型"

```python
# ----------------- 5. 模型评估与可视化 -----------------
# 预测结果
y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]

# 准确率评估
accuracy = accuracy_score(y_test, y_pred)
print(f"\nTest Accuracy: {accuracy:.2f}")
```

#### 步骤2：混淆矩阵解读
"这个矩阵需要重点看两个错误：

- 假阴性（左下角）：将恶性判为良性 → 可能延误治疗

- 假阳性（右上角）：将良性判为恶性 → 造成过度治疗

医疗场景中，假阴性比假阳性更危险！"

```python
# 混淆矩阵热力图
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
           xticklabels=['Benign', 'Malignant'],
           yticklabels=['Benign', 'Malignant'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

#### 步骤3：ROC曲线分析
"这条曲线说明：

- AUC=0.99接近完美 → 模型区分能力极强

- 曲线陡峭上升 → 即便降低阈值（更敏感），也能保持高准确率

- 对比基线（虚线），模型表现非常优秀"

```python
# ROC曲线
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2,
        label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

### 第六部分：特征重要性
"最重要的三个特征：

- worst radius最大半径：癌细胞通常更大

- worst concave points最大凹点：表面凹陷更明显

- mean concave points平均凹点 → 与医生经验一致！

有趣发现：

- fractal dimension分形维度重要性低 → 可能不是关键指标

- smoothness平滑度排名靠后 → 与直观认知不同，需要进一步研究"

```python
# ----------------- 6. 特征重要性分析 -----------------
# 特征重要性排序
importances = rf.feature_importances_  # 获取特征重要性的得分
print(importances)
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12, 8))
plt.title("Feature Importances")
plt.bar(range(X.shape[1]), importances[indices], align="center")
plt.xticks(range(X.shape[1]), data.feature_names[indices], rotation=90)
plt.xlabel("Features")
plt.ylabel("Importance")
plt.tight_layout()
plt.show()
```

### 第七部分：模型解释
"这个病例的预测过程：

- 最大半径显著大于平均值（Z=2.34）

- 纹理异常（Z=-0.56 表示比典型恶性更平滑）

- 模型综合各特征给出87%恶性概率

注意：Z-score解释需要结合业务知识，正值表示高于平均水平"

```python
# ----------------- 7. 模型解释示例 -----------------
# 单样本预测解释
sample_idx = 10  
sample = X_test[sample_idx].reshape(1, -1)
prediction = rf.predict(sample)[0]  # 对样本进行预测
print(prediction)
print(rf.predict(sample))  # array([1])
probability = rf.predict_proba(sample)[0][1]  # 获取样本的恶性概率
print(probability)
print(rf.predict_proba(sample))

print("\nExample Prediction:")
print(f"Predicted class: {'Malignant' if prediction == 1 else 'Benign'}")
print(f"Malignancy probability: {probability:.2f}")
print("Top contributing features:")
for i in indices[:3]:
    print(f"- {data.feature_names[i]}: {sample[0][i]:.2f}")
```
